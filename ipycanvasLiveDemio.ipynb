```python
import ipywidgets as widgets
from ipycanvas import Canvas
from ipyevents import Event
import numpy as np
from PIL import Image, ImageFilter

# Create a canvas
canvas = Canvas(width=280, height=280)
canvas.fill_style = 'black'
canvas.fill_rect(0, 0, canvas.width, canvas.height)
canvas.sync_image_data = True

# Define constants
BRUSH_RADIUS = 10

# Variables
drawing = False
prev_x = None
prev_y = None

# Define event handlers
def on_mouse_down(event):
    global drawing, prev_x, prev_y
    drawing = True
    prev_x, prev_y = event['offsetX'], event['offsetY']

def on_mouse_move(event):
    global drawing, prev_x, prev_y
    if drawing:
        x, y = event['offsetX'], event['offsetY']
        canvas.fill_style = 'white'
        canvas.fill_arc(x, y, BRUSH_RADIUS, 0, 2 * 3.14159)
        if prev_x is not None and prev_y is not None:
            # Draw intermediate points between prev and current points
            dx = x - prev_x
            dy = y - prev_y
            steps = max(abs(dx), abs(dy))
            for i in range(1, steps + 1):
                intermediate_x = prev_x + i * dx / steps
                intermediate_y = prev_y + i * dy / steps
                canvas.fill_arc(intermediate_x, intermediate_y, BRUSH_RADIUS, 0, 2 * 3.14159)
        prev_x, prev_y = x, y

def on_mouse_up(event):
    global drawing
    drawing = False

def clear_canvas(button):
    canvas.fill_style = 'black'
    canvas.fill_rect(0, 0, canvas.width, canvas.height)

def resizeAntiAlias(image_array, new_size):
    old_height, old_width = image_array.shape
    new_height, new_width = new_size

    # Calculate scale factors
    scale_y = old_height / new_height
    scale_x = old_width / new_width

    # Initialize resized image array
    resized_image_array = np.zeros(new_size, dtype=image_array.dtype)

    # Iterate over each pixel in the resized image
    for y in range(new_height):
        for x in range(new_width):
            # Calculate corresponding pixel coordinates in the original image
            src_y = y * scale_y
            src_x = x * scale_x

            # Calculate the surrounding pixel indices
            y0 = int(src_y)
            x0 = int(src_x)
            y1 = min(y0 + 1, old_height - 1)
            x1 = min(x0 + 1, old_width - 1)

            # Calculate interpolation weights
            weight_y = src_y - y0
            weight_x = src_x - x0

            # Perform bilinear interpolation with antialiasing
            top_left = image_array[y0, x0] * (1 - weight_y) * (1 - weight_x)
            top_right = image_array[y0, x1] * (1 - weight_y) * weight_x
            bottom_left = image_array[y1, x0] * weight_y * (1 - weight_x)
            bottom_right = image_array[y1, x1] * weight_y * weight_x
            resized_image_array[y, x] = (top_left + top_right + bottom_left + bottom_right) / 255.0  # Normalize values

    return resized_image_array

def export_image(button):
    # Convert canvas to numpy array
    image_data = canvas.get_image_data(0, 0, canvas.width, canvas.height)
    image_array = np.array(image_data[::-1])  # Vertical flip

    # Convert to grayscale
    if len(image_array.shape) == 3:  # Check if the image has color channels
        image_array = np.mean(image_array, axis=2).astype(np.uint8)  # Convert to grayscale

    # Resize to 28x28 with anti-aliasing
    resized_image_array = resizeAntiAlias(image_array, (28, 28))

    # Threshold to get black/white values
    threshold = 128
    image_array = np.where(resized_image_array < threshold, 0, 255)

    # Display the exported image
    print("Exported Image:")
    print(image_array)

# Setup event listeners
event = Event(source=canvas, watched_events=['mousedown', 'mousemove', 'mouseup'])
event.on_dom_event(on_mouse_down)
event.on_dom_event(on_mouse_move)
event.on_dom_event(on_mouse_up)

# Create buttons
clear_button = widgets.Button(description='Clear Canvas')
export_button = widgets.Button(description='Export Image')

# Add event handlers to buttons
clear_button.on_click(clear_canvas)
export_button.on_click(export_image)

# Display the canvas and buttons
widgets.HBox([canvas, widgets.VBox([clear_button, export_button])])
```
